```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



#Dataset
#https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv
sales = pd.read_csv('company_sales_data.csv') 

sales.head()
```

```python
sales.shape
```

```python
#The columns containing concerned products
product_cols = list(sales.columns[1:7])


month_list= ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec']

#xtick locations for plots
xpos = np.arange(len(sales['month_number']))
```

### Exercise 1; Read Total profit of all months and show it using a line plot

```python
plt.figure(figsize=(10,7))

plt.title('Profit per month')
plt.xlabel('Months',fontsize=13, labelpad=10)
plt.ylabel('Profit',fontsize=13, labelpad=10)
plt.xticks(xpos,month_list)
#plt.yticks()

plt.plot(xpos, sales['total_profit'],marker='o')

plt.grid()
plt.show()
```

### Exercise 2; Read all product sales data and show it  using a multiline plot

```python
plt.figure(figsize=(9,8))

plt.title('Product sales data',fontsize=15)
plt.xlabel('Month',fontsize=13, labelpad=10)
plt.ylabel('Sales unit',fontsize=13, labelpad=10)
plt.xticks(xpos,month_list)

for i in product_cols:
    plt.plot(xpos, sales[i],linewidth=3,marker='o',markersize=7,label= f'{i} Sales')

plt.grid()
plt.legend(loc=0, shadow=True)
plt.show()
```

### Exercise 3: Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure(figsize=(7,5))

plt.title('Toothpaste sales for each month',fontsize=15)
plt.scatter(xpos,sales['toothpaste'], label='Tooth paste sales')

plt.xticks(xpos,month_list)
plt.xlabel('Month number',fontsize=13, labelpad=10)
plt.ylabel('Sales unit',fontsize=13, labelpad=10)


plt.grid()
plt.legend(loc=0,shadow=True)
plt.show()
```

### Exercise 4: Read face cream and facewash product sales data and show it using the bar chart

```python
plt.figure(figsize=(10,8))
plt.title('Face cream and Face wash sales data',fontsize=15)
plt.xticks(xpos,month_list)

plt.xlabel('Month',fontsize=13, labelpad=10)
plt.ylabel('Sales',fontsize=13, labelpad=10)

plt.bar(xpos-0.1,sales['facecream'],width=0.3,label='Face cream')#, align='edge')
plt.bar(xpos+0.2,sales['facewash'],width=0.3,label='Face wash')#, align='edge')

plt.grid()
plt.legend(loc=0,fontsize=12,shadow=True)
plt.show()
```

### Exercise 5: Read sales data of bathing soap of all months and show it using a bar chart.

```python
plt.figure(figsize=(10,8))
plt.title('Bathing soap sales data',fontsize=15)

plt.xlabel('Month',fontsize=13, labelpad=10)
plt.ylabel('Sales Unit',fontsize=13, labelpad=10)
plt.xticks(xpos,month_list)

plt.bar(xpos,sales['bathingsoap'])

plt.grid()
plt.show()
```

### Exercise 6: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.figure(figsize=(10,8))
plt.title('Total profit by month', fontsize=15)
plt.xlabel('Profit Range',fontsize=13, labelpad=10)

plt.hist(sales['total_profit'], bins=5)

#plt.grid()
```

### Exercise 7: Calculate total sale data for last year for each product and show it using a Pie chart
### Note: Display Number of units sold per year for each product in percentage.

```python
total_unit_per_product = [sales[i].sum() for i in product_cols]

plt.pie(total_unit_per_product,labels=[i.upper() for i in product_cols],radius=1.5, autopct='%0.1f%%')

plt.show()
```

### Exercise 8: Read bathing soap and facewash of all months and display it using the Subplot

```python
fig, ax = plt.subplots(2, 1, figsize=(15,8))

ax[0].set_title('Sales data of Bathing soap',fontsize=15)
#ax[0].set_xlabel('Month')
ax[0].set_ylabel('Sales unit',fontsize=13, labelpad=10)
ax[0].plot(month_list,sales['bathingsoap'],marker = 'o',color = 'b')
ax[0].grid()

ax[1].set_title('Sales data of Facewash',fontsize=15)
ax[1].set_xlabel('Month',fontsize=13, labelpad=10)
ax[1].set_ylabel('Sales unit',fontsize=13, labelpad=10)
ax[1].plot(month_list,sales['facewash'],marker = 'o', color = 'r')
ax[1].grid()

plt.subplots_adjust(bottom=0.01)
```

### Exercise 9: Read all product sales data and show it using the stack plot

```python
facecream_Sales = sales['facecream']
facewash_Sales = sales['facewash']
toothPaste_Sales = sales['toothpaste']
bathingsoap_Sales = sales['bathingsoap']
shampoo_Sales = sales['shampoo']
moisturizer_Sales = sales['moisturizer']


plt.figure(figsize=(10,7))
plt.title('Stack Plot for all product', fontsize=15)

for i in product_cols:
    plt.plot([],[], label=i)

plt.xlabel('Month',fontsize=13, labelpad=10)
plt.ylabel('Sales Units',fontsize=13, labelpad=10)
plt.stackplot(month_list, facecream_Sales, facewash_Sales, toothPaste_Sales, bathingsoap_Sales, shampoo_Sales, moisturizer_Sales)

plt.legend(loc=2,shadow=True, fontsize=11)
plt.show()
```
